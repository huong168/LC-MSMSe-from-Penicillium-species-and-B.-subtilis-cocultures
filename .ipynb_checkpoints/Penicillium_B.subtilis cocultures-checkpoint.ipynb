```r
df <- read.csv("7d1733caf7654edda027c323b9b24ca2.csv", header=T)
#df
```

```r
#install.packages("stringr")
library("stringr")
```

```r
colnames(df) <- str_replace_all(colnames(df), '.Peak.area','')
colnames(df) <- str_replace_all(colnames(df), '_Cent','')
```

```r
replace_less_than_100 <- function(x) {
  ifelse(x < 100, 0, x)
}
# Apply the function to all columns except the first three
df[, -c(1:3)] <- apply(df[, -c(1:3)], 2, replace_less_than_100)

# View the modified dataframe
df
```

```r
library(dplyr)

df_fil <- df %>% filter(B_subtilis <= 100)
df_fil
```

```r
all_columns <- colnames(df_fil)

# Separate groups following col_name
group1_columns <- all_columns[grep("^BP\\d{2}$", all_columns)] #group of Confronting zone
group2_columns <- all_columns[grep("^BPs\\d{2}$", all_columns)] #group of Penicillium colonies
group3_columns <- all_columns[grep("^P\\d{2}$", all_columns)] #group of Penicillium axenic

# Filtering data following groups
BP_data <- df_fil[, grepl("^BP\\d{2}$", colnames(df_fil))] #group of Confronting zone
BPs_data <- df_fil[, grepl("^BPs\\d{2}$", colnames(df_fil))]  #group of Penicillium colonies
P_data <- df_fil[, grepl("^P\\d{2}$", colnames(df_fil))] #group of Penicillium axenic
```

```r
BP_data <- cbind(df_fil[1], BP_data)
BPs_data <- cbind(df_fil[1], BPs_data)
P_data <- cbind(df_fil[1], P_data)
#BPs_data
```

```r
# Calculate the average for each ID
mean_BP_data <- rowMeans(BP_data[, -1])  # Exclude the first column (ID)
new_BP_data <- data.frame(ID = unique(df_fil$row.ID), Mean = tapply(mean_BP_data, df_fil$row.ID, mean))

mean_BPs_data <- rowMeans(BPs_data[, -1]) 
new_BPs_data <- data.frame(ID = unique(df_fil$row.ID), Mean = tapply(mean_BPs_data, df_fil$row.ID, mean))

mean_P_data <- rowMeans(P_data[, -1]) 
new_P_data <- data.frame(ID = unique(df_fil$row.ID), P = tapply(mean_P_data, df_fil$row.ID, mean))
#new_P_data
#new_BP_data
#new_BPs_data
```

```r
#combine coculture and axenic
merged_cocult <- merge(new_BP_data, new_BPs_data, by = "ID")
#merged_axenic <- merge(new_P_data, Bsub_data, by = "ID")

# Calculate the mean for each row (by ID) in merged_cocult
mean_cocult_id <- rowMeans(merged_cocult[, -1])
mean_cocult <- data.frame(ID = merged_cocult$ID, Cocult = tapply(mean_cocult_id, merged_cocult$ID, mean))

# Calculate the total for each row (by ID) in merged_axenic
#total_axenic_id <- rowSums(merged_axenic[, -1])
#total_axenic <- data.frame(ID = merged_axenic$ID, Total = tapply(total_axenic_id, merged_axenic$ID, sum))

# View the results
mean_cocult
#total_axenic
```

###


FOLD_CHANGES_WHOLE DATA

###

```r
replace_less_than_1 <- function(x) {
  ifelse(x < 1, 1, x)
}
# Apply the function to all columns except the first three
mean_cocult[-1] <- apply(mean_cocult[-1], 2, replace_less_than_1)
new_P_data[-1] <- apply(new_P_data[-1], 2, replace_less_than_1)
```

```r
# Merge the data frames by ID
merged_cocult_axenic <- merge(mean_cocult, new_P_data, by = "ID", suffixes = c("_mean_cocult", "_new_P_data"))

# Perform the division for each corresponding column
result_cocult_axenic <- merged_cocult_axenic %>%
  mutate(
    Fold_changed = Cocult / P,
  ) %>%
  select(ID,Fold_changed)

result_cocult_axenic
```

```r
write.csv(result_cocult_axenic, file = "result_cocult_Peni_axenic.csv", row.names = TRUE)
```

```r
log_fold_changes <- log10(result_cocult_axenic[-1])
#log_fold_changes
```

```r
log_fold_changes  <- cbind(df_fil[1], log_fold_changes )
#log_fold_changes
```

```r
write.csv(log_fold_changes, file = "log_fold_changes_Penicillium.csv", row.names = TRUE)
```

```r
count_greater_than_1.5 <- result_cocult_axenic %>%
  filter(Fold_changed > 1.5) %>%
  nrow()

count_between_0.5_and_1.5 <- result_cocult_axenic %>%
  filter(Fold_changed >= 0.5 & Fold_changed <= 1.5) %>%
  nrow()

count_less_than_0.5 <- result_cocult_axenic %>%
  filter(Fold_changed < 0.5) %>%
  nrow()

count_nan <- result_cocult_axenic %>%
  filter(is.nan(Fold_changed)) %>%
  nrow()

count_inf <- result_cocult_axenic %>%
  filter(is.infinite(Fold_changed)) %>%
  nrow()

# View the results
print(paste("Count greater than 1.5:", count_greater_than_1.5))
print(paste("Count between 0.5 and 1.5:", count_between_0.5_and_1.5))
print(paste("Count less than 0.5:", count_less_than_0.5))
print(paste("Count of NaN:", count_nan))
print(paste("Count of Inf:", count_inf))
```

```r
replace_less_than_1 <- function(x) {
  ifelse(x < 1, 1, x)
}
# Apply the function to all columns except the first three
new_BP_data[-1] <- apply(new_BP_data[-1], 2, replace_less_than_1)
new_BPs_data[-1] <- apply(new_BPs_data[-1], 2, replace_less_than_1)
new_P_data[-1] <- apply(new_P_data[-1], 2, replace_less_than_1)
```

```r
# Merge the data frames by ID
merged_BP1_axenic <- merge(new_BP_data, new_P_data, by = "ID", suffixes = c("_new_BP_data", "_new_P_data"))
merged_BPs_axenic <- merge(new_BPs_data, new_P_data, by = "ID", suffixes = c("_new_BPs_data", "_new_P_data"))

# Perform the division for each corresponding column
result_BP1_axenic <- merged_BP1_axenic %>%
  mutate(
    Fold_changed = Mean / P,
  ) %>%
  select(ID,Fold_changed)

result_BPs_axenic <- merged_BPs_axenic %>%
  mutate(
    Fold_changed = Mean / P,
  ) %>%
  select(ID,Fold_changed)

result_BP1_axenic
result_BPs_axenic
```

```r
log_fold_changes_BP1 <- log10(result_BP1_axenic[-1])
log_fold_changes_BPs <- log10(result_BPs_axenic[-1])
```

```r
log_fold_changes_BP1 <- cbind(df_fil[1], log_fold_changes_BP1)
log_fold_changes_BPs <- cbind(df_fil[1], log_fold_changes_BPs)
```

```r
#Data log_FC between Confronting zone / Axenic
write.csv(log_fold_changes_BP1, file = "log_fold_changes_BP1_Penicillium.csv", row.names = TRUE)

#Data log_FC between Penicillium zone / Axenic
write.csv(log_fold_changes_BPs, file = "log_fold_changes_BPs_Penicillium.csv", row.names = TRUE)
```
