```r
# Giả sử df là dataframe của bạn

# Lấy tên của tất cả các cột
all_columns <- colnames(df)

# Chia thành 3 nhóm dựa trên tên cột
group1_columns <- all_columns[grep("^BP\\d{2}$", all_columns)]
group2_columns <- all_columns[grep("^BPs\\d{2}$", all_columns)]
group3_columns <- all_columns[grep("^P\\d{2}$", all_columns)]

# Lọc dữ liệu theo nhóm
BP_data <- df[, grepl("^BP\\d{2}$", colnames(df))]
BPs_data <- df[, grepl("^BPs\\d{2}$", colnames(df))]
P_data <- df[, grepl("^P\\d{2}$", colnames(df))]
```

```r
library(ggplot2)
library(dplyr)
```

```r
# Giả sử 'group1_data' và 'group2_data' là hai dataframe của nhóm 1 và nhóm 2

# Chuyển đổi nhóm 1 và nhóm 2 thành ma trận
matrix_BP_data <- as.matrix(BP_data)
matrix_BPs_data <- as.matrix(BPs_data)
matrix_P_data <- as.matrix(P_data)

# Cộng hai ma trận
sum_matrix <- matrix_BP_data + matrix_BPs_data

# Chia kết quả của ma trận cho 2
cocul_matrix <- ((sum_matrix / 2) / matrix_P_data)
cocul_matrix
```

```r
changed_features <- as.data.frame(cocul_matrix)
new_df <- cbind(df[, c(1:3)], changed_features)
new_df
```
